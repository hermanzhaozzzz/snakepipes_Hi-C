# Effective bin size and final ratio

```python
df_resolution = pd.read_csv('../tables/map_resolutions.tsv', sep='\t')
df_resolution = df_resolution.iloc[df_resolution[df_resolution.is_effective.eq('NO')].index - 1,:].query('is_effective!="NO"')
df_resolution = df_resolution.iloc[:,[0,1]].copy()
df_resolution.columns = ['sample', 'valid_bin (Kb)']
df_resolution.set_index('sample', inplace=True)
df_resolution
```

```python
df_pairstats = pd.read_csv('../tables/map_pairstats.tsv', sep='\t', header=None)

# total pe reads
df_pairstats = df_pairstats.loc[df_pairstats[1].eq('Total_pairs_processed'),[0, 2]].copy()
df_pairstats.columns = ['sample', 'counts']
df_pairstats['counts'] = (df_pairstats['counts'] / 1_000_000).map(lambda x: round(x, 2))
df_pairstats.columns = ['sample', 'total_PE (Mb)']
df_pairstats.set_index('sample', inplace=True)
df_pairstats
```

```python
df_valid_interation_stats = pd.read_csv('../tables/valid_interaction_stats.tsv', sep='\t', header=None)
```

```python
df_valid_interation = df_valid_interation_stats.loc[df_valid_interation_stats[1].eq('valid_interaction'), [0, 2]].copy()
df_valid_interation.columns = ['sample', 'counts']
df_valid_interation['counts'] = (df_valid_interation['counts'] / 1_000_000).map(lambda x: round(x, 2))
df_valid_interation.columns = ['sample', 'valid_interation (Mb)']
df_valid_interation.set_index('sample', inplace=True)
df_valid_interation
```

```python
df_valid_interation_rmdup = df_valid_interation_stats.loc[df_valid_interation_stats[1].eq('valid_interaction_rmdup'), [0, 2]].copy()
df_valid_interation_rmdup.columns = ['sample', 'counts']
df_valid_interation_rmdup['counts'] = (df_valid_interation_rmdup['counts'] / 1_000_000).map(lambda x: round(x, 2))
df_valid_interation_rmdup.columns = ['sample', 'valid_interation_rmdup (Mb)']
df_valid_interation_rmdup.set_index('sample', inplace=True)
df_valid_interation_rmdup
```

```python
df_stats = pd.DataFrame([df_resolution.iloc[:,0], df_pairstats.iloc[:,0], df_valid_interation.iloc[:,0], df_valid_interation_rmdup.iloc[:,0]]).T
df_stats
```

```python
df_stats = df_stats[df_stats.index.map(lambda x: x not in ['SRR400264_00', 'SRR400264_01', 'test'])].copy()
df_stats
```

```python
df_stats['% final_ratio'] = (df_stats['valid_interation_rmdup (Mb)'] / df_stats['total_PE (Mb)'] * 100).map(lambda x: round(x, 2))
df_stats.reset_index(inplace=True)
df_stats
```

```python
df_stats['color'] = (df_stats['% final_ratio'] >= 48)
df_stats
```

```python
ggplot(df_stats) + \
    geom_bar(aes(x='sample', y='% final_ratio', group='color', fill='color', color='color'), stat='identity', show_legend=True) + \
    ggsize(600,600) + \
    ggtitle('Final ratio')
```

```python
df_valid_interation_stats.columns = ['sample', 'feature', 'counts']
df_valid_interation_stats = df_valid_interation_stats[df_valid_interation_stats['sample'].map(lambda x: x not in ['SRR400264_00', 'SRR400264_01', 'test'])].copy()
df_valid_interation_stats = df_valid_interation_stats[df_valid_interation_stats['feature'].map(lambda x: x not in ['valid_interaction', 'valid_interaction_rmdup', 'cis_interaction'])].copy()
df_valid_interation_stats
```
